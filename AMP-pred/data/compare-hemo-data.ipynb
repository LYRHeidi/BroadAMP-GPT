```python
import yaml
import pandas as pd
import numpy as np

config = yaml.load(open('/data3/lyr/project_AMP_pre/PeptideBERT-master/config.yaml', 'r'), Loader=yaml.FullLoader)

print(config['task'])
```

```python
print(f'{"="*30}{"DATA":^20}{"="*30}')
with np.load(f'./{config["task"]}/train.npz') as train,\
     np.load(f'./{config["task"]}/val.npz') as val,\
     np.load(f'./{config["task"]}/test.npz') as test:
    train_inputs = train['inputs']
    train_labels = train['labels']
    val_inputs = val['inputs']
    val_labels = val['labels']
    test_inputs = test['inputs']
    test_labels = test['labels']

input_df = pd.concat([pd.DataFrame(train_inputs), pd.DataFrame(val_inputs), pd.DataFrame(test_inputs)])
labels_df = pd.concat([pd.DataFrame(train_labels), pd.DataFrame(val_labels), pd.DataFrame(test_labels)])
input_labels_df = pd.concat([input_df, labels_df], axis=1)
```

```python
m2 = dict(zip(
    range(30), ['[PAD]','[UNK]','[CLS]','[SEP]','[MASK]','L',
    'A','G','V','E','S','I','K','R','D','T','P','N',
    'Q','F','Y','M','H','C','W','X','U','B','Z','O']
))

file='./hemo/train.npz'
f = np.load(file)

# flist = list(map(
#     lambda x: m2[int(x)],
#     list(f['inputs'].flat)
#     ))
flist = np.array(list(map(
    lambda x: m2[int(x)],
    list(f['inputs'].flat)
    ))
    ).reshape(f['inputs'].shape)
```

```python
def combind(flist1):
    my_list = [x for x in flist1 if x != '[PAD]']
    return ''.join(my_list)
combind(flist[1])

flist_combind = []
for i in range(7545):
    flist_combind.append(combind(flist[i]))
```

```python
print(list(f['labels']).count(1))
print(list(f['labels']).count(0))
label_list = list(f['labels'])
a = [x for x, y in list(enumerate(label_list)) if y == 1]
b = [x for x, y in list(enumerate(label_list)) if y == 0]

label_list_b = list(map(
    lambda x: flist_combind[x],
    b
))

label_list_a = list(map(
    lambda x: flist_combind[x],
    a
))
len(label_list_b)
# flist_combind[b[1]]
```
